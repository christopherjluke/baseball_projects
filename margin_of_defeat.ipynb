This scrapes schedules, filters the dataframe by losses, and then creates a run differential column to get the margin of defeat by game.

```python
frames = [ari, atl, bal, bos, chc, chw, cin, cle, col, det, hou, 
          kcr, laa, lad, mia, mil, twins, nym, nyy, oak, phi, pit,
         sdp, sea, sfg, stl, tbr, tex, tor, wsn]

mlb = pd.concat(frames)
mlb = mlb[['Tm', 'RD']]
```

Combine all dataframes and just keep team and run differental.

```python
mlb = mlb.groupby('Tm', as_index=False)['RD'].mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 10))
ax = sns.barplot(data=mlb, 
            y="Tm", 
            x="RD",
            orient='h',
            order = mlb.sort_values('RD').Tm,
            color='#092C5C')
plt.gca().invert_xaxis()
plt.xlabel('Average Margin of Defeat', fontsize=14)
plt.ylabel('Team Names', fontsize=14)
ax.set_title('Margin of Defeat', fontsize=16)
```

```python
pythag = [['LAD', -4],
         ['HOU', 0],
         ['NYM', 4],
         ['ATL', 1],
         ['NYY', -7],
         ['STL', -1],
         ['TBR', 1],
         ['TOR', 2],
         ['SEA', 1],
         ['PHI', 0],
         ['SDP', 4],
         ['MIL', 2],
         ['CLE', 1],
         ['BAL', 3],
         ['CHW', 2],
         ['MIN', -2],
         ['BOS', 1],
         ['SFG', -4],
         ['ARI', -2],
         ['LAA', -4],
         ['COL', 3],
         ['TEX', -8],
         ['CHC', -1],
         ['MIA', -1],
         ['CIN', -2],
         ['KCR', 2],
         ['DET', 2],
         ['PIT', 4],
         ['OAK', 0],
         ['WSN', 1]]

pythag = pd.DataFrame(pythag, columns=['Tm', 'Pythag'])
df = pd.merge(mlb, pythag, how ='inner', on ='Tm')
df.head()
```

Input differential of actual win total to pythag win total into the dataframe with + numbers indicating a win total higher than the pythag win total, and negative numbers indicating an actual win total lower than pythag win total.

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 10))
ax = sns.barplot(data=df, 
            y="Tm", 
            x="RD",
            hue='Pythag',
            dodge=False,
            orient='h',
            order = mlb.sort_values('RD').Tm)
sns.color_palette("colorblind")
plt.gca().invert_xaxis()
plt.xlabel('Average Margin of Defeat', fontsize=14)
plt.ylabel('Team Names', fontsize=14)
ax.set_title('Margin of Defeat', fontsize=16)
```
